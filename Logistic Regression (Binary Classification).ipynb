```python
#import the needed libraries
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#read csv file into dataframe
df= pd.read_csv("HR_comma_sep.csv")
df.head(20)
```

```python
#display dimensions of the dataframe
df.shape
```

```python
#display number of employees who left
df[df.left==1].shape
```

```python
#display number of employees retained
df[df.left==0].shape
```

```python
#group all columns by the column 'left' & display mean of all columns
df.groupby('left').mean()
```

```python
#display impact of salary on employee retention
pd.crosstab(df['salary'], df['left']).plot(kind='bar')
```

```python
#display relationship between employees' departments and their retention rate
pd.crosstab(df['Department'],df['left']).plot(kind='bar')
```

```python
#create a new dataframe that displays only useful columns to be fed into the ML model
subdf= df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
subdf.head()
```

```python
#convert categorical data into numerical data
dummy= pd.get_dummies(subdf['salary'], prefix="salary")
dummy.head()
```

```python
#merge dummy values with the original table
merged= pd.concat([subdf, dummy], axis='columns')
merged.head()
```

```python
#get rid of reference column 'salary'
dropped= merged.drop(["salary"],axis='columns')
dropped.head()
```

```python
#Split data into training data and testing data
X= dropped
y= df.left
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
X_train.head()
```

```python
#create ML model and train it
model= LogisticRegression()
model.fit(X_train, y_train)
```

```python
#predict results of test dataset
y_prediction= model.predict(X_test)
print(y_prediction)
```

```python
#measure accuracy of the model
model.score(X_test,y_test)
```
