```python
#import the needed libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
#read csv file into dataframe
df= pd.read_csv("carprices.csv")
df.head(15)
```

```python
plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')
plt.scatter(df['Mileage'], df['Sell Price($)'], marker=".")
```

```python
#convert categorical data into numerical data
dummy= pd.get_dummies(df['Car Model'])
dummy.head(15)
```

```python
#merge dummy values with the original table
merged= pd.concat([df, dummy], axis='columns')
merged.head(15)
```

```python
#get rid of reference columns
dropped= merged.drop(["Car Model","Mercedez Benz C class"],axis='columns')
dropped.head(15)
```

```python
#display independent variables (input)
X= dropped.drop('Sell Price($)', axis='columns')
X.head(15)
```

```python
#display dependent variable (output)
y= dropped['Sell Price($)']
y.head(15)
```

```python
#create ML model and train it
model= LinearRegression()
model.fit(X,y)
```

```python
#measure accuracy of the model
model.score(X,y)
```

```python
#Predict price of a mercedez benz that is 4 yr old with mileage 45000
model.predict([[45000,4,0,0]])
```

```python
#Predict price of a BMW X5 that is 7 yr old with mileage 86000
model.predict([[86000,7,0,1]])
```
